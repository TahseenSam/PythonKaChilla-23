```python
def rating_conv(x):
    if x.find("p",class_="star-rating One") != None:
        return 1
    elif x.find("p",class_="star-rating Two") != None:
        return 2
    elif x.find("p",class_="star-rating Three") != None:
        return 3
    elif x.find("p",class_="star-rating Four") != None:
        return 4
    elif x.find("p",class_="star-rating Five") != None:
        return 5
    else:
        return 0      
```

```python
print(rating(book[4]))
book[4]
```

```python
for i in book:
    print(i.find("h3").text.strip())
```

```python
book[3]
```

```python
name=[]
rating=[]
price=[]
stock=[]

for i in book:
    name.append(i.find("h3").text.strip())
    rating.append(rating_conv(i))
    price.append(i.find("p", class_="price_color").text.strip())
    stock.append(i.find("p", class_="instock").text.strip())
```

```python
print(name[3])
print(rating[3])
print(price[3])
print(stock[3])
```

# Lets scrap all the book's details

```python
books = soup.find_all("article", class_="product_pod")
len(books)
```

```python
for page in range(1,51):
    url="https://books.toscrape.com/catalogue/page-{0}.html".format(page)
    webpage = requests.get(url).text
    soup = BeautifulSoup(webpage , "lxml")
    books = soup.find_all("article", class_="product_pod")
    for i in books:
        name.append(i.find("h3").text.strip())
        rating.append(rating_conv(i))
        price.append(i.find("p", class_="price_color").text.strip())
        stock.append(i.find("p", class_="instock").text.strip())
```

```python
len(name)
```

```python
print(name[-1])
print(rating[-1])
print(price[-1])
print(stock[-1])
```

```python
x ={
    "Book Name":name,
    "Rating":rating,
    "Price":price,
    "Availaible":stock
}
x
```

```python
pd.DataFrame(x)
```

```python
df=pd.DataFrame(x)
df.to_csv("Books Scraped Data.csv")
```
